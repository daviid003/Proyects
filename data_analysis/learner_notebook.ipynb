```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns  # data visualization library  
import matplotlib.pyplot as plt
import time
import os
import warnings
warnings.filterwarnings("ignore")
```

# Load data

```python
path=os.getcwd()+'\\data.csv'
df = pd.read_csv(path)
df.head()
```

```python
col = df.columns
col
```

```python
y = df.diagnosis
x = df.drop(['Unnamed: 32', 'id', 'diagnosis'], axis=1)
```

```python
x.head()
```

# Visualizing Data

```python
ax = sns.countplot(y, label='Count')
B, M = y.value_counts()
print('Number of Benign Tumors: ',B)
print('Number of Malignant Tumors: ',M)
```

```python
x.describe()
```

```python
df = x
df_std = (df-df.mean())/df.std()
data = pd.concat([y, df_std], axis=1)
data = pd.melt(data, id_vars='diagnosis', var_name='features', value_name='value')
data
```

```python
plt.figure(figsize=(22,15))
sns.violinplot(x='features', y='value', hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45);
plt.grid(True)
plt.ylim([-3,4]);
```

```python
plt.figure(figsize=(25,13))
sns.boxplot(x='features', y='value', hue='diagnosis', data=data)
plt.xticks(rotation=45);
```

# Join plots

```python
sns.jointplot(x.loc[:,'concavity_worst'], x.loc[:,'concave points_worst'], kind='reg', color='#ce1414')
```

# Distributions and Variances with Swarm Plots

```python
sns.set(style='whitegrid', palette='muted')
df = x
df_std = (df-df.mean())/df.std()
data = pd.concat([y, df_std.iloc[:,0:10]], axis=1)
data = pd.melt(data, id_vars='diagnosis', var_name='features', value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features', y='value', hue='diagnosis', data=data);
plt.xticks(rotation=45);
```

```python
sns.set(style='whitegrid', palette='muted')
df = x
df_std = (df-df.mean())/df.std()
data = pd.concat([y, df_std.iloc[:,10:20]], axis=1)
data = pd.melt(data, id_vars='diagnosis', var_name='features', value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features', y='value', hue='diagnosis', data=data);
plt.xticks(rotation=45);
```
